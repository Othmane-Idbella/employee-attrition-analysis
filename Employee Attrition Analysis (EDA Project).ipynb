## Outlier Detection

Outliers are extreme values that fall far outside the typical range of the data.

A boxplot is used to visualize the distribution of employee tenure and detect potential outliers.

```python
# Create a boxplot to visualize the distribution of employee tenure and identify potential outliers
sns.boxplot(data=df1, x='tenure') 
plt.title('Distribution of Employee Tenure')
plt.xlabel('Years at Company')
```

```python
# Calculate outliers using the Interquartile Range (IQR) method
Q1= df1['tenure'].quantile(0.25)
Q3= df1['tenure'].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1 -1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print('Lower limit:', lower_bound)
print('Upper limit:', upper_bound)

outliers = (df1['tenure'] < lower_bound) | (df1['tenure'] > upper_bound)
print('The number of rows containing outliers:', outliers.sum())
```

### Data visualizations

Now, examine variables that you're interested in, and create plots to visualize relationships between variables in the data.

```python
# Visualize the relationship between number of projects, monthly hours, and employee attrition
fig, ax = plt.subplots(1,2, figsize=(22,8))
sns.histplot(data=df1, x='number_of_projects', hue='left', multiple='dodge',shrink=3,ax=ax[1])
ax[1].set_title('Distribution of Number of Projects by Employee Attrition')

sns.boxplot(data=df1, x='average_monthly_hours', y='number_of_projects', hue='left',orient='h', ax=ax[0])
ax[0].invert_yaxis()
```

```python
# Visualize how employee satisfaction levels differ between employees who stayed and those who left
plt.figure(figsize=(10,5))
plt.title('Distribution of Employee Satisfaction by Attrition')
sns.histplot(data=df1, x='satisfaction_level', hue='left', multiple='dodge', bins = 20)
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of Employees")
```

```python
# Analyze how tenure relates to employee satisfaction and attrition
fig, ax = plt.subplots(1,2, figsize=(22,8))

sns.histplot(data=df1, x='tenure', hue= 'left', multiple = 'dodge', shrink = 6, ax =ax[1])
ax[1].set_title('Employee Tenure Distribution by Attrition', fontsize=20)

sns.boxplot(data=df1, x='satisfaction_level', y='tenure', hue = 'left', orient = 'h', ax=ax[0])
ax[0].invert_yaxis()
ax[0].set_title('Employee Satisfaction by Tenure', fontsize=20)
```

```python
# Compare salary distributions between short-tenured and long-tenured employees
short_tenured = df1[df1['tenure']< 7]
long_tenured =df1[df1['tenure']> 6]

fig, ax= plt.subplots(1,2, figsize=(22,8))

sns.histplot(data=short_tenured, x='tenure', hue='salary', hue_order=['low','medium','high'], discrete=1
            ,multiple ='dodge', shrink=.5, ax=ax[0])
ax[0].set_title('Salary Distribution for Short-Tenured Employees')

sns.histplot(data= long_tenured, x= 'tenure', hue='salary', hue_order =['low','medium','high'], discrete =1,
            shrink=.5,multiple='dodge',ax=ax[1])
ax[1].set_title('Salary Distribution for long-tenured employees', fontsize = 20)
```

```python
# Visualize employee attrition across different departments
plt.figure(figsize=(12,8))
sns.histplot(data=df1, x='department', hue='left', multiple='dodge',shrink=.5)
plt.title('Employee Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
```

```python
# Explore the relationship between monthly working hours and recent promotions
plt.figure(figsize=(16,3))
sns.scatterplot(data=df1, x='average_monthly_hours', y='promotion_last_5years',hue='left')
plt.axvline(x=166.6, ls='--',color='r')
plt.title('Relationship Between Monthly Working Hours and Promotions', fontsize=16)
plt.xlabel('Average Monthly Hours')
plt.ylabel('Promotion in Last 5 Years')
```

## Key Insights

Several patterns appear to be associated with employee attrition:

• Employees with lower satisfaction levels are significantly more likely to leave the company.

• Employees working very high monthly hours show higher attrition rates, suggesting possible workload imbalance or burnout.

• Employees assigned a large number of projects appear more likely to leave, indicating that excessive workload may contribute to turnover.

• Promotion opportunities are relatively rare in the dataset, which may contribute to employee dissatisfaction and increased attrition.

• Employees who remain at the company longer than six years appear less likely to leave, suggesting that most attrition occurs earlier in employee tenure.

## Conclusion

This exploratory analysis identified several factors associated with employee attrition at Salifort Motors.

Lower employee satisfaction, high workloads, and limited promotion opportunities appear to contribute to higher turnover rates. Employees managing many projects or working long hours show particularly high attrition levels.

These findings suggest that improving work-life balance, monitoring employee workloads, and increasing promotion opportunities may help the company reduce employee turnover and improve overall retention.
